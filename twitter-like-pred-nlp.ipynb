```python
train=pd.read_csv('/kaggle/input/twitter-likes-prediction-famous/train.csv')
test=pd.read_csv('/kaggle/input/twitter-likes-prediction-famous/test.csv')
sub=pd.read_csv('/kaggle/input/twitter-likes-prediction-famous/sample_submission.csv')
```

```python
import pandas as pd

# Assuming your original DataFrame is named 'original_df'

# Set the random seed for reproducibility (optional)
random_seed = 42
pd.np.random.seed(random_seed)

# Select 2000 random rows from the original DataFrame
sampled_df = train.sample(n=2000)

# Create a new DataFrame from the sampled rows
new_df = pd.DataFrame(sampled_df)

# Reset the index of the new DataFrame
new_df.reset_index(drop=True, inplace=True)
```

```python
train=new_df
```

```python
X=train['tweets']
y=train['likes']

test=test['tweet']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Tokenize the questions
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X)

test=vectorizer.transform(test)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error

# Instantiate the models
rf_reg = RandomForestRegressor(random_state=42,verbose =2)
# gb_reg = GradientBoostingRegressor(random_state=42)
# cat_reg = CatBoostRegressor(random_state=42, verbose=0)

# Train the models
rf_reg.fit(X_train, y_train)
# gb_reg.fit(X_train, y_train)
# cat_reg.fit(X_train, y_train)

# Make predictions on the test set
rf_pred = rf_reg.predict(X_test)
# gb_pred = gb_reg.predict(X_test)
# cat_pred = cat_reg.predict(X_test)

# Calculate the mean squared error for each model
rf_mse = mean_squared_error(y_test, rf_pred)
# gb_mse = mean_squared_error(y_test, gb_pred)
# cat_mse = mean_squared_error(y_test, cat_pred)

print("Random Forest Regressor MSE: {:.4f}".format(rf_mse))
# print("Gradient Boosting Regressor MSE: {:.4f}".format(gb_mse))
# print("CatBoost Regressor MSE: {:.4f}".format(cat_mse))
```

```python
y_pred=rf_reg.predict(test)
```

```python
import joblib

# Save the pre-trained model
joblib.dump(rf_reg, 'twitter_prediction.pkl')

# Save the CountVectorizer
joblib.dump(vectorizer, 'vectorizer.pkl')
```
